## 问题 1

根据下列递归公式, 计算下列 $T(n)$ 对应的的渐近上界。要求所求的边界尽可能的紧（tight）, 请写明步骤。

1. $T(1)=1; T(n)=T(n/4)+1$ for $n>1$
2. $T(1)=1;T(n)=3T(n/3)+n^2$ for $n>1$
3. $T(1)=1;T(n)=T(2n/3)+1$ for $n>1$
4. $T(1)=1;T(n)=5T(n/4)+n$ for $n>1$
5. $T(n)=1 \ for\ n \le 2 ; T(n)=T(\sqrt{n})+1 \ for \ n>2$


1. $T(n)=T(n/4)+1$

展开递推：
$$
T(n)=T(n/4^k)+k
$$

当 $n/4^k=1$ 时结束，所以 $4^k=n$，即：
$$
k=\log_4 n
$$

因此：
$$
T(n)=1+\log_4 n=\Theta(\log n)
$$

2. $T(n)=3T(n/3)+n^2$

由 Master Theorem：
$$
a=3,\quad b=3,\quad f(n)=n^2
$$

$$
n^{\log_b a}=n^{\log_3 3}=n
$$

因为 $f(n)=n^2$ 比 $n$ 多一个多项式因子，且：
$$
3f(n/3)=3(n/3)^2=n^2/3 \leq c n^2
$$

其中 $c<1$，所以：
$$
T(n)=\Theta(n^2)
$$

3. $T(n)=T(2n/3)+1$

展开递推：
$$
T(n)=T((2/3)^k n)+k
$$

当 $(2/3)^k n=1$ 时结束，所以：
$$
k=\log_{3/2} n
$$

因此：
$$
T(n)=1+\log_{3/2} n=\Theta(\log n)
$$

4. $T(n)=5T(n/4)+n$

由 Master Theorem：
$$
a=5,\quad b=4,\quad f(n)=n
$$

$$
n^{\log_b a}=n^{\log_4 5}
$$

因为：
$$
\log_4 5>1
$$

所以：
$$
f(n)=n=O(n^{\log_4 5-\epsilon})
$$

因此：
$$
T(n)=\Theta(n^{\log_4 5})
$$

5. $T(n)=T(\sqrt n)+1$

令：
$$
n=2^m
$$

则：
$$
\sqrt n=2^{m/2}
$$

设：
$$
S(m)=T(2^m)
$$

原递推变为：
$$
S(m)=S(m/2)+1
$$

展开：
$$
S(m)=S(m/2^k)+k
$$

当 $m/2^k \leq 1$ 时结束，所以：
$$
k=\log_2 m
$$

又因为：
$$
m=\log_2 n
$$

所以：
$$
T(n)=\Theta(\log \log n)
$$


## 问题 2

观察下方的分治算法（divide-and-conquer algorithm）的伪代码, 回答下面问题

```latex
DoSomething(A,p,r)
-----
n := r-p+1
if n=2 and A[p]>A[r] then
    swap A[p] and A[r]
else if n >= 3 then
    m = ceil(2n/3)
    DoSomething(A,p,p+m-1)
    DoSomething(A,r-m+1,r)
    DoSomething(A,p,p+m-1)  
    
---
first call: DoSomething(A,1,n)
```

note：$ceil(2n/3)=\left\lceil {2n/3} \right\rceil$；$:=$ 表示赋值, 等价于 $\to$；A是一个包含n的整数元素的数组, 

1. 写出该算法时间复杂度的递归公式, 并求解其对应的渐进表示
2. 描述一下该算法的功能, 并判断是否是最高效的解决方案
3. \* 使用python编程实现上述算法或其对应的更高效版本的算法

idea:

该算法是 Stooge Sort 的一种形式，功能是将数组按非递减顺序排序。

设输入规模为 $n$。当 $n=1$ 或 $n=2$ 时，只需要常数时间；当 $n \geq 3$ 时，算法会递归调用 3 次，每次处理约 $2n/3$ 个元素，因此时间复杂度递推式为：

$$
T(n)=3T(\lceil 2n/3\rceil)+O(1)
$$

忽略取整后：

$$
T(n)=3T(2n/3)+O(1)
$$

由 Master Theorem：

$$
a=3,\quad b=3/2,\quad f(n)=O(1)
$$

因此：

$$
T(n)=\Theta(n^{\log_{3/2}3})
$$

其中：

$$
\log_{3/2}3 \approx 2.71
$$

所以该算法的时间复杂度为：

$$
T(n)=\Theta(n^{2.71})
$$

该算法可以完成排序，但不是高效算法。普通比较排序可以达到 $O(n\log n)$，例如 merge sort、heap sort；Python 内置排序一般也是 $O(n\log n)$。因此该算法主要用于理解递归和分治，不适合实际排序。


````python
def do_something(A, p, r):
    n = r - p + 1

    if n == 2 and A[p] > A[r]:
        A[p], A[r] = A[r], A[p]
    elif n >= 3:
        m = (2 * n + 2) // 3   # ceil(2n/3)
        do_something(A, p, p + m - 1)
        do_something(A, r - m + 1, r)
        do_something(A, p, p + m - 1)

    return A


A = [5, 2, 9, 1, 6, 3]
do_something(A, 0, len(A) - 1)
print(A)

# time complexity: T(n)=3T(ceil(2n/3))+O(1)=Theta(n^(log_{3/2}3))≈Theta(n^2.71)
````

## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:

这里的 `profits[i]` 表示第 `i` 天的收益，所以问题可以转化为：寻找一个连续子数组，使得这个连续区间的和最大。该连续区间的起点就是买入日期，终点就是卖出日期。

一种直接方法是枚举所有买入日期和卖出日期，计算每个区间的收益总和，取最大值。这个方法时间复杂度为 $O(n^2)$。

更高效的方法是使用 Kadane 算法。遍历数组时维护一个当前连续区间收益 `current_sum`。如果之前的连续收益已经小于 0，那么继续加下去只会拖累之后的收益，因此从当前天重新开始计算。遍历过程中不断更新最大收益以及对应的买入、卖出日期。

时间复杂度：

$$
O(n)
$$

空间复杂度：

$$
O(1)
$$


````python
def max_stock_profit(profits):
    if not profits:
        return 0, None, None

    max_sum = profits[0]
    current_sum = profits[0]

    start = 0
    best_start = 0
    best_end = 0

    for i in range(1, len(profits)):
        if current_sum < 0:
            current_sum = profits[i]
            start = i
        else:
            current_sum += profits[i]

        if current_sum > max_sum:
            max_sum = current_sum
            best_start = start
            best_end = i

    return max_sum, best_start + 1, best_end + 1


profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]

max_profit, buy_day, sell_day = max_stock_profit(profits)

print("最大收益:", max_profit)
print("买入日期: 第", buy_day, "天")
print("卖出日期: 第", sell_day, "天")

# time complexity: O(n)
# space complexity: O(1)
````

## 问题 4

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

idea:

因为数组元素互不相同，并且满足 $A[1]>A[2]$、$A[n-1]<A[n]$，所以数组内部一定存在至少一个局部最小值。

要在 $O(\log n)$ 时间内找到一个局部最小值，可以使用二分思想。每次检查中间位置 `mid`：

如果：

$$
A[mid] < A[mid-1]
$$

并且：

$$
A[mid] < A[mid+1]
$$

那么 `A[mid]` 就是局部最小值。

如果 `A[mid] > A[mid-1]`，说明左边相邻元素更小，那么左侧一定存在局部最小值，因此向左搜索。

如果 `A[mid] > A[mid+1]`，说明右边相邻元素更小，那么右侧一定存在局部最小值，因此向右搜索。

每次搜索范围减半，所以时间复杂度为：

$$
O(\log n)
$$

空间复杂度为：

$$
O(1)
$$

如果要找出所有局部最小值，则必须检查每个内部元素，因为局部最小值可能有多个，输出本身也可能很多。因此可以遍历数组中第 2 个到第 n-1 个元素，判断它是否比左右邻居都小。

找出所有局部最小值的时间复杂度为：

$$
O(n)
$$


````python
def find_one_local_min(A):
    left = 1
    right = len(A) - 2

    while left <= right:
        mid = (left + right) // 2

        if A[mid] < A[mid - 1] and A[mid] < A[mid + 1]:
            return A[mid]

        if A[mid] > A[mid - 1]:
            right = mid - 1
        else:
            left = mid + 1

    return None


def find_all_local_min(A):
    result = []

    for i in range(1, len(A) - 1):
        if A[i] < A[i - 1] and A[i] < A[i + 1]:
            result.append(A[i])

    return result


A = [9, 3, 7, 2, 1, 4, 5]

print("一个局部最小值:", find_one_local_min(A))
print("所有局部最小值:", find_all_local_min(A))

# find_one_local_min time complexity: O(log n)
# find_all_local_min time complexity: O(n)
````

## 问题 5

给定包含n个不同数字的一组数, 寻找一种基于比较的算法在这组数中找到k个最小的数字, 并按顺序输出它们。

1. 将n个数先进行排序, 然后按顺序输出最小的k个数。要求：选择合适的排序算法实现上述操作, 计算算法时间复杂度, 并使用python编程实现。
2. 建立一个包含这n个数的堆（heap）, 并且调用 k 次Extract-min 按顺序输出最小的k个数。使用往空堆中不断插入元素的方法建立堆, 分析这种方法建堆的时间复杂度, 并使用python编程实现
3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

idea:

方法 1：先排序，再输出前 k 个元素。可以使用归并排序，归并排序是基于比较的排序算法，时间复杂度为：

$$
O(n\log n)
$$

排序完成后输出前 k 个元素需要：

$$
O(k)
$$

所以总时间复杂度为：

$$
O(n\log n+k)=O(n\log n)
$$

方法 2：建立最小堆，然后连续执行 k 次 Extract-min。题目要求使用不断向空堆插入元素的方法建堆。第 i 次插入的时间复杂度为 $O(\log i)$，所以建堆总时间复杂度为：

$$
\sum_{i=1}^{n}O(\log i)=O(n\log n)
$$

之后执行 k 次 Extract-min，每次时间复杂度为 $O(\log n)$，所以总时间复杂度为：

$$
O(n\log n+k\log n)
$$

方法 3：当数据量太大，无法全部存入内存时，可以只维护一个大小为 k 的最大堆。依次读取数据，如果堆中元素不足 k 个，就直接加入；如果堆已经有 k 个元素，就将当前元素和堆顶最大值比较。若当前元素更小，则删除堆顶，并插入当前元素。这样扫描完所有数据后，堆中保留的就是前 k 小的数。

该方法只需要存储 k 个元素，空间复杂度为：

$$
O(k)
$$

每个元素最多进行一次堆操作，堆大小始终不超过 k，因此时间复杂度为：

$$
O(n\log k)
$$

最后将 k 个元素排序输出，需要：

$$
O(k\log k)
$$


````python
import heapq


def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    result = []
    i = 0
    j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result


def k_smallest_by_sort(arr, k):
    sorted_arr = merge_sort(arr)
    return sorted_arr[:k]


def k_smallest_by_min_heap(arr, k):
    heap = []

    for x in arr:
        heapq.heappush(heap, x)

    result = []
    for _ in range(k):
        result.append(heapq.heappop(heap))

    return result


def k_smallest_large_data(data, k):
    max_heap = []

    for x in data:
        if len(max_heap) < k:
            heapq.heappush(max_heap, -x)
        elif x < -max_heap[0]:
            heapq.heapreplace(max_heap, -x)

    result = [-x for x in max_heap]
    result.sort()
    return result


arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3

print("方法1，排序:", k_smallest_by_sort(arr, k))
print("方法2，最小堆:", k_smallest_by_min_heap(arr, k))
print("方法3，大数据最大堆:", k_smallest_large_data(arr, k))

# method 1 time complexity: O(n log n)
# method 2 time complexity: O(n log n + k log n)
# method 3 time complexity: O(n log k + k log k), space complexity: O(k)
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:

只有一种情况会发生碰撞：左边的星球向右运动，右边的星球向左运动，即栈顶元素为正数、当前元素为负数。

可以使用栈来模拟。依次扫描数组中的星球：

如果当前星球不会和栈顶星球碰撞，就直接入栈。

如果栈顶星球为正数，当前星球为负数，则比较它们的质量：

- 如果栈顶质量小于当前星球质量，栈顶星球消失，当前星球继续和前面的星球比较。
- 如果栈顶质量大于当前星球质量，当前星球消失。
- 如果两者质量相同，根据题意，自右向左运动的星球消失，所以当前负数星球消失，栈顶正数星球保留。

每个星球最多入栈一次、出栈一次，所以时间复杂度为：

$$
O(n)
$$

空间复杂度为：

$$
O(n)
$$


````python
def planet_collision(A):
    stack = []

    for planet in A:
        alive = True

        while alive and stack and stack[-1] > 0 and planet < 0:
            if abs(stack[-1]) < abs(planet):
                stack.pop()
            elif abs(stack[-1]) > abs(planet):
                alive = False
            else:
                # 质量相同：自右向左运动的星球消失，正数星球保留
                alive = False

        if alive:
            stack.append(planet)

    return stack


A1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
A2 = [23, -8, 9, -3, -7, 9, -23, 22]

print(planet_collision(A1))
print(planet_collision(A2))

# time complexity: O(n)
# space complexity: O(n)
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

这是一个 0-1 背包问题。每个大理石只能选择装入或不装入，不能拆分。

设 `dp[i][w]` 表示只考虑前 `i` 个大理石、背包容量为 `w` 时能够获得的最大价值。

对于第 `i` 个大理石，重量为 `weight`，价值为 `value`：

如果不装入第 `i` 个大理石：

$$
dp[i][w]=dp[i-1][w]
$$

如果装入第 `i` 个大理石：

$$
dp[i][w]=dp[i-1][w-weight]+value
$$

因此状态转移方程为：

$$
dp[i][w]=\max(dp[i-1][w], dp[i-1][w-weight]+value)
$$

当 `weight > w` 时，不能装入，只能有：

$$
dp[i][w]=dp[i-1][w]
$$

如果有 `n` 个物品，背包容量为 `W`，时间复杂度为：

$$
O(nW)
$$

空间复杂度为：

$$
O(nW)
$$


````python
def knapsack_01(items, capacity):
    n = len(items)

    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        weight, value = items[i - 1]

        for w in range(capacity + 1):
            if weight > w:
                dp[i][w] = dp[i - 1][w]
            else:
                dp[i][w] = max(
                    dp[i - 1][w],
                    dp[i - 1][w - weight] + value
                )

    return dp[n][capacity]


items = [(5, 10), (4, 40), (6, 30), (3, 50)]
capacity = 9

max_value = knapsack_01(items, capacity)

print("最大价值:", max_value)

# time complexity: O(nW)
# space complexity: O(nW)
````

输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

idea:

二叉搜索树的中序遍历结果应该是一个递增序列。如果恰好有两个节点的值被交换，那么中序遍历序列中会出现逆序现象。

例如：

$$
[1,3,2,4]
$$

其中 `3 > 2`，说明 `3` 和 `2` 的位置有问题。

如果两个被交换的节点不相邻，可能会出现两次逆序。例如：

$$
[3,2,1]
$$

第一次逆序是 `3 > 2`，第二次逆序是 `2 > 1`，真正需要交换的是第一次逆序中的前一个节点 `3`，以及第二次逆序中的后一个节点 `1`。

因此可以对二叉搜索树进行中序遍历，记录上一个访问的节点 `prev`。当发现 `prev.val > current.val` 时：

- 第一次发现逆序时，记录 `first = prev`
- 每次发现逆序时，更新 `second = current`

最后交换 `first.val` 和 `second.val`，即可在不改变树结构的情况下恢复二叉搜索树。

时间复杂度为：

$$
O(n)
$$

因为每个节点只访问一次。

空间复杂度为：

$$
O(h)
$$

其中 `h` 是树的高度，来自中序遍历使用的栈。


````python
from collections import deque


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_tree(values):
    if not values or values[0] is None:
        return None

    root = TreeNode(values[0])
    queue = deque([root])
    i = 1

    while queue and i < len(values):
        node = queue.popleft()

        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1

        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1

    return root


def tree_to_list(root):
    if root is None:
        return []

    result = []
    queue = deque([root])

    while queue:
        node = queue.popleft()

        if node is None:
            result.append(None)
        else:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)

    while result and result[-1] is None:
        result.pop()

    return result


def recover_tree(root):
    stack = []
    current = root
    prev = None
    first = None
    second = None

    while stack or current:
        while current:
            stack.append(current)
            current = current.left

        current = stack.pop()

        if prev and prev.val > current.val:
            if first is None:
                first = prev
            second = current

        prev = current
        current = current.right

    if first and second:
        first.val, second.val = second.val, first.val

    return root


root1 = build_tree([1, 3, None, None, 2])
recover_tree(root1)
print(tree_to_list(root1))

root2 = build_tree([3, 1, 4, None, None, 2])
recover_tree(root2)
print(tree_to_list(root2))

# time complexity: O(n)
# space complexity: O(h)
````

## 问题 10

给你一个二叉树的根节点 root ，树中每个节点都存放有一个 0 到 9 之间的数字。每条从根节点到叶节点的路径都代表一个数字：

> 例如，从根节点到叶节点的路径 1 -> 2 -> 3 表示数字 123 。

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

e.g.:
---
1.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-1.jpg"> </div>

 输入：root = [1,2,3], 输出：25

解释：从根到叶子节点路径 1->2 代表数字 12，1->3 代表数字 13，因此和为 12 + 13 = 25 。

2.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-2.jpg"> </div>

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

idea:

可以使用深度优先搜索 DFS。

从根节点开始向下遍历，每经过一个节点，就把当前路径形成的数字更新为：

$$
current = current \times 10 + node.val
$$

当遍历到叶节点时，说明已经形成了一个完整的根到叶子的数字，将它加入答案。

例如路径 `4 -> 9 -> 5`：

$$
((4 \times 10 + 9) \times 10 + 5)=495
$$

每个节点只访问一次，所以时间复杂度为：

$$
O(n)
$$

递归调用栈的深度等于树的高度，所以空间复杂度为：

$$
O(h)
$$


````python
from collections import deque


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_tree(values):
    if not values or values[0] is None:
        return None

    root = TreeNode(values[0])
    queue = deque([root])
    i = 1

    while queue and i < len(values):
        node = queue.popleft()

        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1

        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1

    return root


def sum_numbers(root):
    def dfs(node, current):
        if node is None:
            return 0

        current = current * 10 + node.val

        if node.left is None and node.right is None:
            return current

        return dfs(node.left, current) + dfs(node.right, current)

    return dfs(root, 0)


root1 = build_tree([1, 2, 3])
root2 = build_tree([4, 9, 0, 5, 1])

print(sum_numbers(root1))
print(sum_numbers(root2))

# time complexity: O(n)
# space complexity: O(h)
````
